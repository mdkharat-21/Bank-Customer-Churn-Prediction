```python
#1)problem statement:
#It is the dataset of a U.S. bank customer for getting the information that , this particular customer will leave bank or not.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#2)Collect the data
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
#3)EDA-(Exploratory Data Analysis):
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#Remove the columns which are not important
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
sns.boxplot(x=df['CreditScore'])
```

```python
q1 = df['CreditScore'].quantile(0.25)
q3 = df['CreditScore'].quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
```

```python
df['CreditScore'] = np.clip(df['CreditScore'], lower, upper)
```

```python
sns.boxplot(df['CreditScore'])
```

```python
sns.boxplot(x=df['Age'])
```

```python
sns.boxplot(x=np.sqrt(df['Age']))
```

```python
sns.boxplot(x=df['Tenure'])
```

```python
sns.boxplot(x=df['Balance'])
```

```python
sns.boxplot(x=df['NumOfProducts'])
```

```python
sns.boxplot(x=np.log(df['NumOfProducts']))
```

```python
sns.boxplot(x=df['EstimatedSalary'])
```

```python
#4)Feature Engineering
```

```python
df.info
```

```python
df = pd.get_dummies(df, columns=['Geography'], dtype=int)
```

```python
 df['Gender'].unique()
```

```python
df.replace({'Female':0, 'Male':1},inplace=True)
```

```python
df.info
```

```python
#5)Feature Selection
```

```python
plt.figure(figsize=(10,5))
df.corr()
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
#6)Model Building
#1)Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
#7)Model Evalution
y_pred=log_model.predict(x_train)
```

```python
cnf=confusion_matrix(y_train,y_pred)
print(cnf)
```

```python
clf=classification_report(y_train,y_pred)
print(clf)
```

```python
y_pred=log_model.predict(x_test)
```

```python
cnf=confusion_matrix(y_test,y_pred)
print(cnf)
```

```python
clf=classification_report(y_test,y_pred)
print(clf)
```

```python

acc_score=accuracy_score(y_test,y_pred)
print(acc_score)
```
